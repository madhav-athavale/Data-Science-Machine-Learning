```python
%matplotlib inline
import numpy as np
from pylab import *
import yfinance as yf
from scipy import stats
```

I have used Kimberly Clark(Ticker :KMB) and Proctor and Gamble (Ticker: PG) to decide whether there is any relationsip between these two stocks. Both of them are competitors in several markets

```python
def find_r_between_ticker_and_spx_index(ticker,start_date,end_date):
    yf_ticker = yf.Tickers([ticker,'^GSPC'])
    yf_data = yf_ticker.history(start=start_date, end=end_date, interval="1d")
    
    slope, intercept, r_value, p_value, std_err = stats.linregress(yf_data['Close','^GSPC'], yf_data['Close',ticker])
    fig = plt.figure(figsize=(16,6))
    plt.scatter(yf_data['Close','^GSPC'],yf_data['Close',ticker])
    plt.plot(yf_data['Close', '^GSPC'], slope*yf_data['Close','^GSPC'] + intercept, c='r')
    plt.xlabel('S & P 500 Index',fontsize=16, c='b')
    plt.ylabel('META',fontsize=16, c='r')    
    plt.show()
    return (slope, intercept, r_value, p_value, std_err)
```

```python
ret = find_r_between_ticker_and_spx_index('MSFT',"2020-01-01", "2023-01-26")

print(ret)
```
